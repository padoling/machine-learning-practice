```python
import pandas as pd
```

```python
path = 'C:/Users/books/Downloads/all/train.csv'
```

```python
titanic = pd.read_csv(path)
```

```python
titanic
```

```python
titanic = titanic.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'])
```

```python
titanic
```

```python
titanic.count()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy='most_frequent')
```

```python
titanic[['Embarked']] = imputer.fit_transform(titanic[['Embarked']])
```

```python
titanic.count()
```

```python
imputer.set_params(strategy='mean')
```

```python
titanic[['Age']] = imputer.fit_transform(titanic[['Age']])
```

```python
titanic.count()
```

```python
titanic
```

```python
sex_df = pd.get_dummies(titanic[['Sex']])
```

```python
sex_df
```

```python
emb_df = pd.get_dummies(titanic[['Embarked']])
```

```python
emb_df
```

```python
titanic = pd.concat((titanic.drop(columns=['Sex', 'Embarked']),sex_df, emb_df), axis=1)
```

```python
titanic
```

```python
train = titanic.sample(frac=0.8)
valid = titanic.drop(index=train.index)
```

```python
train
```

```python
valid
```

```python
train_x = train.drop(columns=['Survived'])
train_y = train[['Survived']]
valid_x = valid.drop(columns=['Survived'])
valid_y = valid[['Survived']]
```

```python
import keras
```

```python
from keras.models import Sequential
```

```python
nn_model = Sequential()
```

```python
from keras.layers import InputLayer, Dense
```

```python
nn_model.add(InputLayer(input_shape=(10,)))
```

```python
nn_model.summary()
```

```python
nn_model.add(Dense(units=10, activation='sigmoid'))
```

```python
nn_model.add(Dense(units=6, activation='sigmoid'))
```

```python
nn_model.add(Dense(units=3, activation='sigmoid'))
```

```python
nn_model.add(Dense(units=1, activation='sigmoid'))
```

```python
nn_model.summary()
```

```python
nn_model.compile(optimizer='sgd', loss='mean_squared_error', metrics=['accuracy'])
```

```python
nn_model.fit(x=train_x, y=train_y, epochs=1000, 
             validation_data=(valid_x, valid_y))
```

```python
from keras.regularizers import l2
```

```python
nn_model3 = Sequential()
```

```python
nn_model3.add(InputLayer((10, )))

nn_model3.add(Dense(units=10, activation='relu', 
                    kernel_regularizer=l2(), bias_regularizer=l2()))
nn_model3.add(Dense(units=6, activation='relu', kernel_regularizer=l2(), bias_regularizer=l2()))
nn_model3.add(Dense(units=3, activation='relu', kernel_regularizer=l2(), bias_regularizer=l2()))

nn_model3.add(Dense(units=1, activation='sigmoid', kernel_regularizer=l2(), bias_regularizer=l2()))

nn_model3.compile(optimizer='sgd', loss='mean_squared_error', metrics=['accuracy'])
```

```python
nn_model3.summary()
```

```python
nn_model3.fit(x=train_x, y=train_y, epochs=200,
              validation_data=(valid_x, valid_y))
```

```python
from keras.layers import Dropout
```

```python
nn_model4 = Sequential()

nn_model4.add(InputLayer((10, )))

nn_model4.add(Dense(units=10, activation='relu'))
nn_model4.add(Dropout(rate=0.2))
nn_model4.add(Dense(units=6, activation='relu'))
nn_model4.add(Dropout(rate=0.2))
nn_model4.add(Dense(units=3, activation='relu'))
nn_model4.add(Dropout(rate=0.2))

nn_model4.add(Dense(units=1, activation='sigmoid'))

nn_model4.compile(optimizer='sgd', loss='mean_squared_error', metrics=['accuracy'])
```

```python
nn_model4.fit(x=train_x, y=train_y, epochs=500,
              validation_data=(valid_x, valid_y))
```

```python
nn_model4.evaluate(x=train_x, y=train_y)
```

```python
nn_model4.evaluate(x=valid_x, y=valid_y)
```

```python
nn_model5 = Sequential()

nn_model5.add(InputLayer((10, )))

nn_model5.add(Dense(units=10, activation='relu'))
nn_model5.add(Dropout(rate=0.2))
nn_model5.add(Dense(units=6, activation='relu'))
nn_model5.add(Dropout(rate=0.2))
nn_model5.add(Dense(units=3, activation='relu'))
nn_model5.add(Dropout(rate=0.2))

nn_model5.add(Dense(units=9, activation='softmax'))

nn_model5.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
from keras.optimizers import SGD, Adam
```

```python
adam_opt = Adam(lr=0.02, beta_1= 0.5, beta_2=0.9)
```

```python
nn_model5.compile(optimizer=adam_opt, loss='mean_squared_error', metrics=['accuracy'])
```

```python
nn_model5.fit(x=train_x, y=train_y, batch_size=16, epochs=70,
              validation_data=(valid_x, valid_y))
```
