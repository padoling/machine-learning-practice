1. Deep Learning이란?
    - machine lenarning의 일종.<br>
    데이터를 통해 컴퓨터가 '학습'을 하고 그것을 사용해 컴퓨터가 자동으로 문제를 해결하도록 하는 것.<br>
    이 때 프로그램에 적용하는 rule은 원래 사람이 만들어서 줬었음.<br>
    그런데 사람이 rule을 만들 때에는 어려움이 존재하기 때문에 data와 정답으로 기계를 학습시켜(machine learning) rule을 만들어 프로그램에 적용시킴.<br>

    - Perceptron: 뇌의 신경세포를 모델링해서 만든 것.그러나 문제 많았음.<br>

    - deep learning이 다시 뜬 이유

        - 빅 데이터(데이터의 수가 많아짐)

### Improve models
1. Activation Function: Single perceptron에 나타난 결과를 어떻게 변형시킬 것인가.
    - Binary activation: 특정 역치값을 넘으면 활성화, 안넘으면 비활성화. 그러나 수학적인 관점에서 보면 이러한 activation함수를 가지고 활성화를 하는 것에 문제가 있음. 미분값이 항상 0이기 때문에 고정된 값을 이용해 update를 수행해야 함.
    - Sigmoid activation: 0과 1 사이의 실수 값을 가짐. 미분이 굉장히 쉬우며 binary activation과 유사한 역할을 함.
    - Tanh(탄젠트) Activation: Sigmoid와 유사하지만 -1과 1 사이의 값을 가짐. 미분이 간단하고 sigmoid보다 미분값이 더 크기 때문에 weight값이 더 빠르게 수정됨.<br>
-> Vanishing gradient problem: layer가 깊어질 수록 gradient값이 점점 작아짐. output layer에서 멀어질수록 weight값이 거의 업데이트되지 않음. sigmoid를 여러 개 겹치다보면 미분값이 점점 0으로 수렴하게 됨.
    - ReLU activation: 0보다 클 때는 y=ax에 따르고, 0보다 작을 때는 무조건 0. 즉, 0보다 작으면 training을 하지 않고, 0보다 크면 고정된 값(a)으로 training을 함. 그러나 0일 때 미분이 불가능하다는 단점이 있으며, 초기 weight값 지정에 많은 영향을 받음.
<br>
<br>
2. Overfit and Underfit: 두 경우에 빠지지 않았을 경우를 general한 경우라고 일컬음.
    - SVM: 두 class 사이에 선을 그을 때, 각 class와의 간격(gap)이 최대가 되도록 선을 긋는 방법. deep learning 이전에 generalization 관점에서 매우 유용했던 모델.
    - Regularization: 모델에 일부러 패널티를 부여. error를 계산할 때 x들의 기울기에 따른 패널티를 줘서 모델을 좀 더 general하게 만듦.
    - Dropout: 랜덤하게 일부 node를 제외하고 학습시키는 모델. 
        - Co-adaption(지나치게 모델에 적합화되는 것)회피: 영향력이 강한 노드를 제외시켜서 다른 노드도 학습할 수 있는 기회를 줌.
        - Ensemble효과(여러 개의 모델을 합쳐서 하나로 만드는 것): 각 batch에 따라 학습된 여러 모델을 합쳐서 하나의 모델로 만듦. 각 batch는 서로 다른 dropout을 진행한 모델들.
<br>
<br>
3. Optimizer
    - SGD: Stochastic Gradient Descent(경사 하강법). 매 epcoh(학습)마다 error값을 계산하고, error의 미분값에 따라 error값이 점점 작아지는 형태로 모델을 수정하는 기법. minimum값이 될 때까지 진행.<br>
    -> 지역 최소값(local minima)에 빠지는 문제점이 발생.
        - momentum optimizer: 가속도값을 이용해서 원래 이동하려는 방향으로 계속 나아가려는 성질을 통해, 얕은 local minima를 빠져나갈 수 있도록 학습.
        - Various optimizer: Adagrad, Rmsprop, Adam 등.
        - batch training: training할 때 데이터를 하나씩 넣어서 학습시킬 수 있고, 전체 데이터를 넣어서 학습시킬 수 있음. batch training은 데이터의 일부분만 떼어서 training을 하는 방법. local minima에 빠질 가능성이 적음.

```python
from keras.regularizers import l2
```

```python
nn_model3 = Sequential()
```

```python
nn_model3.add(InputLayer((10,)))

nn_model3.add(Dense(units=10, activation='relu', 
                    kernel_regularizer=l2(), bias_regularizer=l2()))    #regularizer 추가. l2() 안에 가중치를 부여할 수 있음.
nn_model3.add(Dense(units=6, activation='relu', kernel_regularizer=l2(), bias_regularizer=l2()))
nn_model3.add(Dense(units=3, activation='relu', kernel_regularizer=l2(), bias_regularizer=l2()))

nn_model3.add(Dense(units=1, activation='sigmoid'))

nn_model3.compile(optimizer='sgd', loss='mean_squared_error', metrics=['accuracy'])
```

```python
nn_model3.summary()
```

```python
nn_model3.fit(x=train_x, y=train_y, epochs=200,
             validation_data=(valid_x, valid_y))    #학습 실행
```

조교님 모델에선 학습률 더 안좋게 나옴. l2의 값이 너무 커서 underfitting됐다고 보면 됨. 그럴 경우 l값을 줄여주면 됨.<br>
근데 사실 최근엔 regularization 잘 쓰지 않음 dropout만 써도 거의 잘 학습됨.<br>
Dropout은 현재 많이 쓰고 있는 기법.

```python
from keras.layers import Dropout
```

```python
nn_model4 = Sequential()

nn_model4.add(InputLayer((10,)))

nn_model4.add(Dense(units=10, activation='relu'))
nn_model4.add(Dropout(rate=0.2))   #rate: 몇%만큼 dropout시킬것인가. 여기서는 hidden unit이 10개이므로 2개를 dropuout시키겠다는 말.
#node의 수가 많을 땐 보통 0.5정도로 잡는다고 함.
nn_model4.add(Dense(units=6, activation='relu'))
nn_model4.add(Dropout(rate=0.2))
nn_model4.add(Dense(units=3, activation='relu'))
nn_model4.add(Dropout(rate=0.2))

nn_model4.add(Dense(units=1, activation='sigmoid'))  #ouput layer에는 dropout 적용시키지 않음.

nn_model4.compile(optimizer='sgd', loss='mean_squared_error', metrics=['accuracy'])
```

```python
nn_model4.fit(x=train_x, y=train_y, epochs=500,
             validation_data=(valid_x, valid_y))    #학습 실행
```

dropout을 쓰면 training과 validation의 갭이 좀 크게 나옴.

```python
nn_model4.evaluate(x=train_x, y=train_y)   #얼마나 제대로 학습이 됐는지 보는 것. 결과에서 앞이 loss, 뒤가 accuracy 정보.
```

```python
nn_model4.evaluate(x=valid_x, y=valid_y)
```

evaluate해서 나온 결과가 실제 모델의 결과. train 결과와 valid 결과의 값이 좀 더 가까워짐.<br>
마지막으로 optimizer를 바꿔보자.

```python
nn_model5 = Sequential()

nn_model5.add(InputLayer((10,)))

nn_model5.add(Dense(units=10, activation='relu'))
nn_model5.add(Dropout(rate=0.2))   
nn_model5.add(Dense(units=6, activation='relu'))
nn_model5.add(Dropout(rate=0.2))
nn_model5.add(Dense(units=3, activation='relu'))
nn_model5.add(Dropout(rate=0.2))

nn_model5.add(Dense(units=1, activation='sigmoid'))  

nn_model5.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])   #optimizer만 adam으로 변경
```

```python
nn_model5.fit(x=train_x, y=train_y, epochs=70, batch_size=16,
             validation_data=(valid_x, valid_y))
```

1 epoch: 전체 데이터를 한 번 학습 시키는 것.<br>
batch: 전체 데이터 중 일부만 가져와서 여러 번 학습 시킴. <br>
예를 들어 전체 데이터 개수가 1000개이고 batch size가 100이라면 100개의 데이터를 10번 학습함. 10번 학습한 것이 1 epoch.<br>
batch_size를 16으로 설정하면 1 epoch마다 약 50번씩 학습하는 것이므로 epoch 수를 좀 줄여줘도 됨.

학습이 좀 더 빠르게 진행됨. 약 200번 정도 학습한 때에서 결과가 고정됨. loss값이 error인데 마지막으로 가면 error값이 증가하는 모습을 보임. overfitting이 일어나고 있다는 의미.
