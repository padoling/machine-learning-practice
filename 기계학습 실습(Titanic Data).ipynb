다음으로 데이터를 분석하는 데에 필요없는 열들을 삭제한다.<br>
Id code나 이름, ticket과 같은 overfitting을 유발할 만한 정보는 쓸모 없으므로 삭제한다.<br>
행이나 열을 통째로 삭제할 땐 drop 메서드를 쓰면 된다.<br>
열 정보를 삭제하려면 columns = '열 이름'의 형태로 써 주어야 한다.

```python
titanic = titanic.drop(columns = 'PassengerId')
```

```python
titanic = titanic.drop(columns = 'Name')
```

```python
titanic = titanic.drop(columns = 'Ticket')
```

세 개의 열을 삭제한 후의 정보이다.<br>
간단하게 각 열 별로 몇 개의 valid한 값이 존재하는지 확인하려면 count()함수를 쓰면 된다.

```python
titanic.count()
```

확인해보니 Cabin 열의 값의 개수가 204개밖에 되지 않는다.<br>
이는 결측값의 수가 매우 많다는 의미이므로, Cabin 열도 삭제해주자.

```python
titanic = titanic.drop(columns = 'Cabin')
```

이제 남은 열들에서 결측값이 있는 행을 삭제해준다.<br>
dropna 함수는 결측값이 있는 행이나 열을 통째로 삭제해준다.<br>
이 때 axis의 값을 0으로 설정하면 행, 1로 설정하면 열을 나타낸다.<br>
how의 값을 'any'로 설정하면 해당 행에 하나라도 결측값이 있을 경우 해당 행을 삭제하며,<br>
'all'로 설정하면 해당 행의 모든 값이 결측값일 경우 행을 삭제한다.

```python
titanic = titanic.dropna(axis=0, how='any')
```

```python
titanic.count()
```

```python
titanic
```

이제 전체 데이터를 training set과 validation set으로 분리해보자.<br>
sample()함수는 dataframe에서 행을 기준으로 한 random sample을 반환한다.<br>
frac 값은 어느 정도의 양을 분리해낼 것인지를 나타낸다.<br>
여기서는 training set을 0.8, 즉 80%만큼 분리해내보자.

```python
train = titanic.sample(frac=0.8)
```

다음과 같이 trian에 들어갈 instance 570개가 무작위로 선택된 것을 볼 수 있다.

## 전처리
sklearn을 이용하여 데이터 분석 이전에 전처리를 시행한다.<br>
sklearn.preprocessing에 다양한 전처리 기법들이 존재한다.<br>
먼저 nominal data인 Sex와 Embarked를 numeric data로 바꿔준다.<br>
두 가지 방법이 존재하는데, OneHotEncoder와 OrdinalEncoder이다.

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
```

OrdinalEncoder는 각 값에 대해 0부터 숫자를 부여하는 방식이다.<br>
Sex의 경우 Female = 0, Male = 1과 같은 식으로 인코딩될 수 있다.<br>
fit(): Train the model using libsvm

```python
ord_enc = OrdinalEncoder()
```

```python
ord_enc.fit(train_x[['Sex', 'Embarked']])
```

학습시킨 ord_enc가 정보를 어떻게 분류했는지 categories_를 통해 살펴볼 수 있다.<br>
Sex의 경우 female, male 순으로, Embarked의 경우 C,Q,S 순으로 분류되었다.<br>
즉, female에 0이, male에 1이 대입되었다는 말이다.

```python
ord_enc.categories_
```

```python
train_x[['Sex', 'Embarked']] = ord_enc.transform(train_x[['Sex', 'Embarked']])
```

```python
train_x
```

그런데 train과 valid가 random하게 나뉘어졌기 때문에 category가 다르게 지정이 될 수 있다.<br>

```python
OrdinalEncoder(categories=[['male', 'female'],['C', 'S', 'Q']])
```

이번에는 OneHotEncoder를 이용해보자.<br>
OneHotEncoder는 각 카테고리의 수 만큼 행렬을 만들어 1 또는 0을 부여하는 방식이다.<br>
예를 들어, Sex의 경우 (male, female)이라는 일종의 벡터 형태를 지정해 값이 male인 경우 (1,0)으로 기술하는 식이다.<br>
반대로 값이 female인 경우에는 (0,1)이 될 것이다.

OneHotEncoder는 Sex와 Embarked에 따로 적용시켜주자.<br>
categories에 변환할 카테고리를 직접 지정해주면 된다.

```python
onehot_sex_enc = OneHotEncoder(categories=[['male', 'female']])
```

fit_transform은 fit과 transform을 한꺼번에 해 줄 수 있다.
toarray()를 이용해 array형태로 만들어줄 수 있다.

```python
onehot_sex_enc.fit_transform(train_x[['Sex']]).toarray()
```

```python
onehot_sex_enc.categories_
```

이제 train_x 데이터에 변환한 Sex를 덮어씌워야 하는데, 에러가 난다.<br>
그 이유는 하나였던 Sex column이 OneHot을 이용하면서 female, male의 두 개의 column으로 늘어났기 때문이다.

```python
train_x[['Sex']] = onehot_sex_enc.fit_transform(train_x[['Sex']]).toarray()
```

먼저 index가 뒤죽박죽이므로 0부터 다시 매겨준다. reset_index()를 사용하면 됨.<br>
그런데 이걸 쓰면 index라는 열이 새로 생기므로 drop시켜준다.

```python
train_x = train_x.reset_index()
```

```python
train_x = train_x.drop(columns=['index'])
```
