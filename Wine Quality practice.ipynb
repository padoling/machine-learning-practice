### 결측값 제거

```python
wine = wine.dropna(axis=0, how='any')
```

```python
wine.count()
```

### 전처리
- OneHotEncoder 사용해서 type을 numeric한 값으로 바꿔줌.

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
onehot_type_enc = OneHotEncoder(categories = [['red', 'white']])
```

```python
onehot_type_enc.fit_transform(wine[['type']]).toarray()
```

```python
onehot_type_enc.categories_
```

```python
type_df = pd.DataFrame(data=onehot_type_enc.fit_transform(wine[['type']]).toarray(),
                      columns=['red','white'])
```

```python
wine = wine.reset_index()
wine = wine.drop(columns=['index'])
```

```python
wine = pd.concat((type_df, wine.drop(columns=['type'])), axis=1)
```

```python
wine
```

wine = wine[wine.quality != 3]
wine.count()

## quality 값 삭제

```python
wine = wine[wine.quality != 8]
wine = wine[wine.quality != 9]
wine['quality'].value_counts()
```

## test, valid, train set으로 구분하기

```python
test = wine.sample(frac=0.2)
```

```python
valid = wine.drop(index=test.index).sample(frac=0.25)
```

```python
train = wine.drop(index=test.index)
train = train.drop(index=valid.index)
```

### 독립변수와 종속변수 구분(x, y)

```python
train_x = train.drop(columns = 'quality')
train_y = train[['quality']]
```

```python
valid_x = valid.drop(columns='quality')
valid_y = valid[['quality']]
test_x = test.drop(columns='quality')
test_y = test[['quality']]
```
