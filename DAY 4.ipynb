```python
import pandas as pd
```

```python
titanic_path = "C:/Users/books/Downloads/all/train.csv"
```

```python
titanic = pd.read_csv(titanic_path)
```

```python
titanic = titanic.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'])
```

```python
titanic.Sex = titanic.Sex.astype('category')
titanic.Embarked = titanic.Embarked.astype('category')
```

```python
titanic.count()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy='most_frequent')
```

```python
titanic[['Embarked']] = imputer.fit_transform(titanic[['Embarked']])
```

```python
imputer.set_params(strategy='mean')
```

```python
titanic[['Age']] = imputer.fit_transform(titanic[['Age']])
```

```python
sex_df = pd.get_dummies(titanic[['Sex']])
```

```python
emb_df = pd.get_dummies(titanic[['Embarked']])
```

```python
titanic = pd.concat((titanic.drop(columns=['Sex', 'Embarked']), sex_df, emb_df), axis=1)
```

```python
titanic
```

```python
train = titanic.sample(frac=0.8)
valid = titanic.drop(index=train.index)
train_y = train[['Survived']].values
train_x = train.drop(columns=['Survived']).values
valid_y = valid[['Survived']].values
valid_x = valid.drop(columns=['Survived']).values
```

```python
import keras
```

```python
from keras.models import Sequential
```

```python
model = Sequential()
```

```python
from keras.layers import Dense, InputLayer
```

```python
model.add(InputLayer(input_shape=(10,)))
```

```python
model.summary()
```

```python
model.add(Dense(units=100, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.add(Dense(units=50, activation='sigmoid'))
```

```python
model.add(Dense(units=20, activation='sigmoid'))
```

```python
model.add(Dense(units=10, activation='sigmoid'))
```

```python
model.add(Dense(units=1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error', metrics=['accuracy'])
```

```python
model.fit(x=train_x, y=train_y, batch_size=10, epochs=1000, validation_data=(valid_x, valid_y))
```
